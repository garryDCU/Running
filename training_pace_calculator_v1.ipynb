### Training Calculator Ideas

First draft of a training calculator

```python
result=240
pace_diff=4
```

```python
lap=float(result)/3.75
eight=lap-pace_diff
three_k=lap+pace_diff
five_k=lap+(2*pace_diff)
ten_k=lap+(3*pace_diff)
threshold=lap+(4*pace_diff)

print("""Current Training Paces:
800m = {}, 
1500m = {},
3000m = {}, 
5000m = {},  
10,000m = {}, 
threshold = {}""".format(eight,lap,three_k,five_k,ten_k,threshold))
```

```python
# Specific work with duration equal to race length 
# Can change in future to increase difficulty of sessions etc

# 800m pace work with recovery equal to 4x interval duration in seconds
pace8_short = 0.5*float(eight)
rec8_short = 4*pace8_short
pace8_long = float(eight)
rec8_long = 4*pace8_long

print("""
800m Specific Session Options:
SETS OF...
a) 4x 200m in {} with {} seconds walk/jog recovery
b) 2x 400m in {} with {} seconds walk/jog recovery""".format(pace8_short, rec8_short, pace8_long, rec8_long))

# 1500m pace work with recovery equal to 2x interval duration in seconds
pace15_short = float(lap)*3/4
rec15_short = 2*pace15_short
pace15_long = float(lap)*5/4
rec15_long = 2*pace15_long

print("""
1500m Specific Session Options:
SETS OF...
a) 5x 300m in {} with {} seconds walk/jog recovery
b) 3x 500m in {} with {} seconds walk/jog recovery""".format(pace15_short, rec15_short, pace15_long, rec15_long))

# 3000m pace work with recovery equal to 1x interval duration in seconds
pace3_short = float(three_k)*3/4
rec3_short = pace3_short
pace3_long = float(three_k)*3/2
rec3_long = pace3_long

print("""
3000m Specific Session Options:
SETS OF...
a) 10x 300m in {} with {} seconds walk/jog recovery
b) 5x 600m in {} with {} seconds walk/jog recovery""".format(pace3_short, rec3_short, pace3_long, rec3_long))

```

```python
# Threshold Options 

total_duration=1800
```

```python
# CONTINUOUS PROGRESSIVE TEMPO (if less than 30 mins or doing sub_max)

if total_duration <= 1800:
  total_dur=total_duration/60
  l_pace=threshold*2.5
  min_l_pace=(l_pace//60)
  sec_l_pace=l_pace-((l_pace//60)*60)
  u_pace=l_pace+10
  min_u_pace=(u_pace//60)
  sec_u_pace=u_pace-((u_pace//60)*60)


  session1=("Continuous {} minutes starting at {}:{} per km and progressing to {}:{} per km".format(int(total_dur), int(min_u_pace), sec_u_pace, int(min_l_pace), sec_l_pace))

else:
  total_dur=total_duration/60
  l_pace=threshold*2.5
  u_pace=l_pace+10
  u_plus_pace=l_pace+20

  min_u_plus_pace=(u_plus_pace//60)
  sec_u_plus_pace=u_plus_pace-((u_plus_pace//60)*60)
  min_u_pace=(u_pace//60)
  sec_u_pace=u_pace-((u_pace//60)*60)


  session1=("Continuous {} minutes starting at {}:{} per km and progressing to {}:{} per km".format(int(total_dur), int(min_u_plus_pace), sec_u_plus_pace, int(min_u_pace), sec_u_pace))  

# THRESHOLD REPS (if more than 30 mins)
rep_duration=600

n_reps=total_duration/rep_duration
rep_dur=rep_duration/60
min_rep_pace=int((threshold*2.5)//60)
sec_rep_pace=(threshold*2.5)-(((threshold*2.5)//60)*60)

rep_rest=rep_duration*0.2
min_rep_rest=(rep_rest//60)
sec_rep_rest=(rep_rest)-(rep_rest//60)*60


if sec_rep_rest!=0:
  session2=("""{} x {} minute reps in {}:{} per km with {}:{} rest""".format(int(n_reps), int(rep_dur), int(min_rep_pace), sec_rep_pace, int(min_rep_rest), int(sec_rep_rest)))
else:
  if sec_rep_pace!=0:
    session2=("""{} x {} minute reps in {}:{} per km with {} minutes rest""".format(int(n_reps), int(rep_dur), int(min_rep_pace), sec_rep_pace, int(min_rep_rest)))
  else:
    session2=("""{} x {} minute reps in {} minutes per km with {} minutes rest""".format(int(n_reps), int(rep_dur), int(min_rep_pace), int(min_rep_rest)))

if total_duration<2400:
  print("""
  THRESHOLD SESSION OPTIONS
  a) {} 
  b) {} 
  """.format(session1,session2))
else:
  print("""
  THRESHOLD SESSION OPTIONS
  {}
  """.format(session2))
```
